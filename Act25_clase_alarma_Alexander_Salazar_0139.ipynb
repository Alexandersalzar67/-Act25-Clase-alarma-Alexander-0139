```python
class Alarma:
    def __init__(self, tipo, estado, ubicacion):
        self.tipo = tipo
        self.estado = estado
        self.ubicacion = ubicacion

    def leerdatos(self, nuevo_tipo, nuevo_estado, nueva_ubicacion):
        self.tipo = nuevo_tipo
        self.estado = nuevo_estado
        self.ubicacion = nueva_ubicacion
        print(f"Datos actualizados en la ubicación: {self.ubicacion}")

    def mostrardatos(self):
        print(f"Configuración de Alarma -> Tipo: {self.tipo}, Estado: {self.estado}, Ubicación: {self.ubicacion}")

# Uso de la clase
alarma1 = Alarma("Incendio", "Activada", "Cocina")
print(f"Alarma 1: {alarma1.tipo}, Ubicación: {alarma1.ubicacion}")
alarma1.mostrardatos()

print("-" * 20)

alarma2 = Alarma("Robo", "Desactivada", "Entrada Principal")
print(f"Alarma 2: {alarma2.tipo}, Ubicación: {alarma2.ubicacion}")
alarma2.mostrardatos()
# Ejemplo de uso de la función leerdatos para actualizar
alarma2.leerdatos("Movimiento", "Activada", "Garaje")
alarma2.mostrardatos()

print("\ncreador de software Alexander Salazar NL = 49")
```
